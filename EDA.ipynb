```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv(
    r"C:\Users\workk\OneDrive\Desktop\churn prediction project\WA_Fn-UseC_-Telco-Customer-Churn (1).csv"
)

df.head()
```

```python
print("Shape of Dataset:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Churn',
    data=df
)

plt.title("Churn Distribution")

plt.show()
```

```python
churn_percent = (
    df['Churn']
    .value_counts(normalize=True)
    * 100
)

print(churn_percent)
```

```python
numerical_cols = [

    'tenure',

    'MonthlyCharges',

    'TotalCharges'
]

for col in numerical_cols:

    plt.figure(figsize=(7,4))

    sns.histplot(
        df[col],
        kde=True
    )

    plt.title(f'Distribution of {col}')

    plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Churn',
    y='tenure',
    data=df
)

plt.title("Churn vs Tenure")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Churn',
    y='MonthlyCharges',
    data=df
)

plt.title("Churn vs Monthly Charges")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Contract',
    hue='Churn',
    data=df
)

plt.title("Contract Type vs Churn")

plt.xticks(rotation=15)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    x='PaymentMethod',
    hue='Churn',
    data=df
)

plt.title("Payment Method vs Churn")

plt.xticks(rotation=25)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='InternetService',
    hue='Churn',
    data=df
)

plt.title("Internet Service vs Churn")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='SeniorCitizen',
    hue='Churn',
    data=df
)

plt.title("Senior Citizen vs Churn")

plt.show()
```

```python
df_corr = df.copy()

df_corr["Churn"] = df_corr["Churn"].map({
    "Yes":1,
    "No":0
})
```

```python
df_corr = pd.get_dummies(
    df_corr,
    drop_first=True
)
```

```python
df_corr = pd.get_dummies(
    df_corr,
    drop_first=True
)
```

```python
corr_matrix = df_corr.corr()
```

```python
plt.figure(figsize=(16,10))

sns.heatmap(
    corr_matrix,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
churn_corr = (
    corr_matrix["Churn"]
    .sort_values(ascending=False)
)

print(churn_corr.head(15))
```
